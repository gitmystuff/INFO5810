# Week 4 - Make Regression

```python
# deterministic
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

X, y = make_regression(n_samples=100, n_features=1, noise=0)
plt.scatter(X, y)
plt.show();
```

```python
# probabilistic
X, y = make_regression(n_samples=100, n_features=1, noise=50)
plt.scatter(X, y)
plt.show();
```

```python
# make_regression to dataframe
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

X, y = make_regression(n_samples=100, n_features=3, n_targets=1, noise=50)
df = pd.DataFrame(X, columns=['f1', 'f2', 'f3'])
df['y'] = y
print(df.info())
print(df.head())

sns.pairplot(df, x_vars=['f1','f2','f3'], y_vars='y', 
             kind='reg', 
             height=5,
             aspect=0.8, 
             plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.2}});
```

```python
# make_regression to dataframe
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

X, y = make_regression(n_samples=100, n_features=3, n_targets=1, n_informative=2, noise=50)
df = pd.DataFrame(X, columns=['f1', 'f2', 'f3'])
df['y'] = y
print(df.head())

sns.pairplot(df, x_vars=['f1','f2','f3'], y_vars='y', 
             kind='reg', 
             height=5,
             aspect=0.8, 
             plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.2}});
```
