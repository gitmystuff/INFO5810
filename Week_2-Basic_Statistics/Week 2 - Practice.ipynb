```python
# getting data from sklearn dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
print(housing.keys())
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
X_train, X_test, y_train, y_test = train_test_split(df.drop('MedHouseVal', axis=1), df['MedHouseVal'], test_size=.2, random_state=42)
print(X_train.head())
print(X_train.info())
print(X_train.describe())
# if you just need X and y
# X, y = fetch_california_housing(return_X_y=True)
```

```python
# getting data from seaborn datasets
import seaborn as sns
from sklearn.model_selection import train_test_split

mpg = sns.load_dataset('mpg')
X_train, X_test, y_train, y_test = train_test_split(mpg.drop('mpg', axis=1), mpg['mpg'], test_size=.2, random_state=42)
print(X_train.head())
print(X_train.info())
print(X_train.describe())
```

```python
# getting data from a file
import pandas as pd
from sklearn.model_selection import train_test_split

advertising = pd.read_csv('Advertising.csv', usecols=['TV', 'radio', 'newspaper', 'sales'])
X_train, X_test, y_train, y_test = train_test_split(advertising.drop('sales', axis=1), advertising['sales'], test_size=.2, random_state=42)
print(X_train.head())
print(X_train.info())
print(X_train.describe())
```
